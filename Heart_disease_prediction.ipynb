```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# loading the csv data to a Pandas DataFrame
heart_data = pd.read_csv('heart_disease_data.csv')
```

```python
# print first 5 rows of the dataset
heart_data.head()
```

```python
# print last 5 rows of the dataset
heart_data.tail()
```

```python
# number of rows and columns in the dataset
heart_data.shape
```

```python
# getting some info about the data
heart_data.info()
```

```python
# checking for missing values
heart_data.isnull().sum()
```

```python
# statistical measures about the data
heart_data.describe()
```

```python
# checking the distribution of Target Variable
heart_data['target'].value_counts()
```

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model training

```python
model = LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

## Model evalution

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

```python
#model performance
cm = confusion_matrix(Y_train, X_train_prediction)
cm
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Plot confusion matrix using seaborn heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix on train dataste')
plt.xlabel('Predicted Label') 
plt.ylabel('True Label') 
plt.show() 
```

```python

from sklearn.metrics import classification_report

# Generate classification report

report = classification_report(Y_train, X_train_prediction)

# Print the report 
print("Classification Report:")

print(report) 
```

```python
#model performance on test dataset 
cm2 = confusion_matrix(X_test_prediction, Y_test)
cm2 
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 

# Plot confusion matrix using seaborn heatmap

plt.figure(figsize=(6, 4))
sns.heatmap(cm2, annot=True, fmt='d', cmap='Blues')

plt.title('Confusion Matrix on test dataset')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show() 
```

```python
# Generate classification report
from sklearn.metrics import classification_report
report = classification_report(X_test_prediction, Y_test)

# Print the report
print("Classification Report:")
print(report) 
```
